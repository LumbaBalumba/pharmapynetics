```python
from pharmapynetics.models import EPBFTPK
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
t = np.array([0.,  0.5,  1.,  1.5,  2.,  3.,  3.33,  3.66,  4.,
              4.33,  4.66,  5.,  5.33,  5.66,  6.,  7.,  8., 10.,
              12., 16., 24., 48., 72., 96.])
X = np.array([0.,   98.3, 1150., 1340., 1480., 1400., 1350., 1440.,
              1360., 1340., 1180., 1100., 1080., 1090., 1060., 1030.,
              990.,  964., 1030.,  811.,  597.,  302.,  145.,   59.])
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
t_max = 25
```

```python
t_current = t[t <= t_max]
X_current = X[t <= t_max]
```

```python
model = EPBFTPK(n_models=3, l=10.0, clipped=True, t_max=t_max)
```

```python
model.fit(t_current, X_current)
```

```python
X_pred = model.sample(t)
```

```python
plt.plot(t, X)
plt.plot(t, X_pred)
```
